```python
import math
import matplotlib.pyplot as plt
```

```python
def insertBadJob(b):
    bad_rb_cost = 0
    threshold = 0
    curr_b = 0
    while(curr_b<b):
        for i in range(0,2**threshold):
            bad_rb_cost = bad_rb_cost+(2**threshold)
            curr_b=curr_b+1
        threshold=threshold+1
    return bad_rb_cost
```

```python
def bounceGoodJob(b,M):
    good_rb_cost = 0
    algo_cost = 0
    for j in range(0,int(math.log(b,2))+1):
        good_rb_cost = good_rb_cost + (j+1)*M*(2**j)
    algo_cost = good_rb_cost + (((math.log(b+1,2))+1)*M) + b
    return algo_cost
```

```python
def addAlphaBadJob(algo_cost, bad_rb_cost, alpha):
    algo_cost = algo_cost+alpha
    bad_rb_cost = bad_rb_cost+alpha
    return algo_cost,bad_rb_cost
```

```python
def makePlot(algo_cost,bad_rb_cost):
    plt.loglog(bad_rb_cost,algo_cost,'orange',marker='s')
    plt.axline([1000, 1000], [10000000, 10000000],linestyle ='--')
```

```python
def main(M_exp,alpha,iteration):
    b_cost=[]
    g_cost=[]
    bad_rb_cost = 0
    algo_cost = 0
    for x in range(10,28):
        result=[]
        bad_rb_cost = insertBadJob((2**x-1))
        algo_cost = bounceGoodJob((2**x),(2**M_exp))
        result = addAlphaBadJob(algo_cost,bad_rb_cost,0)
        g_cost.append(result[0])
        b_cost.append(result[1])
    for i in range(0,len(g_cost)):
        g_cost[i] = g_cost[i]*iteration
        b_cost[i] = b_cost[i]*iteration
    makePlot(g_cost,b_cost)
    print('algo_cost',g_cost)
    print('b_cost',b_cost)
```

```python
main(1,0,10)
```
