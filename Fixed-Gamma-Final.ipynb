```python
import math
import matplotlib.pyplot as plt
import numpy as np
gamma = 1
n=[256, 1024, 2048, 3072, 4096, 5120, 6016, 7040, 8064, 9088, 10112, 11008, 12928]
```

```python
def createG(N,k,a):
    g_array= [0]*N
    for index in range(0,len(g_array)):
        if index%k >= (k-a):
            g_array[index]=1
    return g_array
```

```python
def createHatG(N,k,a):
    hatg_array= [0]*N
    for index in range(0,len(hatg_array)):
        if index%(k) < a:
            hatg_array[index]=1
    return hatg_array
```

```python
def count_goodJob(array):
    count=0
    for c in array:
        if c==1:
            count=count+1
    return count
```

```python
def calculateCost(hatg_array,g_array):
    algoCost=0
    badCost=0
    index=1
    count_bad=0
    for job in hatg_array:
        if job == 1:
            algoCost=algoCost+pow(index,gamma)
            index=1
        else:
            badCost=badCost+pow(index,gamma)
            index=index+1
    
    for job in g_array:
        if job == 0:
            count_bad = count_bad+1
    algoCost = algoCost+count_bad
    return algoCost,badCost
```

```python
def makePlot(A_1,B_1,A_2,B_2,A_3,B_3,A_4,B_4,A_5,B_5):

    
    plt.loglog(B_1,A_1,'red',marker='.')
    plt.loglog(B_2,A_2,'teal',marker='o')
    plt.loglog(B_3,A_3,'orange',marker='s')
    plt.loglog(B_4,A_4,'black',marker='*')
    plt.loglog(B_5,A_5,'green',marker='v')

    #plt.axline([10^3, 10^3], [10^5, 10^5],linestyle ='--')
    #plt.xlim([10000, 1000000000000])
    #plt.ylim([1000, 10000000])

    plt.legend(['alpha=1','alpha=8','alpha=16','alpha=32','alpha=64'])
    plt.xlabel('Total Bad cost')
    plt.ylabel('Total Algorithm cost')
    plt.title('Simulation for gamma = 1')
    plt.show()
```

```python
def calculate():
    alpha=[1,8,16,32,4] 
    A_1=[0]*(len(n))
    B_1=[0]*(len(n))
    A_2=[0]*(len(n))
    B_2=[0]*(len(n))
    A_3=[0]*(len(n))
    B_3=[0]*(len(n))
    A_4=[0]*(len(n))
    B_4=[0]*(len(n))
    A_5=[0]*(len(n))
    B_5=[0]*(len(n))
    Good=[0]*(len(n))
    for a in alpha:
        print('For alpha = ',a)
        i=0
        for N in n:
            result=[]
            hatg_array = createHatG(N,128,a)
            g_array = createG(N,128,a)
            Good[i] = count_goodJob(g_array)
            print('Experiment with',N, 'jobs')
            #print('Fix hat{g}',hatg_array)
            #print('Fix g',g_array)
            result = calculateCost(hatg_array,g_array)
            if a==alpha[0]:
                A_1[i] = result[0]
                B_1[i] = result[1]
            if a==alpha[1]:
                A_2[i] = result[0]
                B_2[i] = result[1]
            if a==alpha[2]:
                A_3[i] = result[0]
                B_3[i] = result[1]
            if a==alpha[3]:
                A_4[i] = result[0]
                B_4[i] = result[1]
            if a==alpha[4]:
                A_5[i] = result[0]
                B_5[i] = result[1]
            i=i+1
    print('A_1:',A_1)
    print('B_1:',B_1)
    print('A_2:',A_2)
    print('B_2:',B_2)
    print('A_3:',A_3)
    print('B_3:',B_3)
    print('A_4:',A_4)
    print('B_4:',B_4)
    print('A_5:',A_5)
    print('B_5:',B_5)
    makePlot(A_1,B_1,A_2,B_2,A_3,B_3,A_4,B_4,A_5,B_5)
```
